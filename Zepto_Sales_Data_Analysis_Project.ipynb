```python
# Convert 'order_date' to datetime objects
df_sales["order_date"] = pd.to_datetime(df_sales["order_date"])
```

## **Data Analysis - Aggregations and Grouping**

```python
df_sales.columns
```

```python
min_amount = df_sales["total_amount"].min()
print(f"Min Total Amount: {min_amount:.2f}")
```

```python
max_amount = df_sales["total_amount"].max()
print(f"Max Total Amount: {max_amount:.2f}")
```

```python
avg_amount = df_sales["total_amount"].mean()
print(f"Avrage of  Total Amount: {avg_amount:.2f}")
```

```python
df_sales.columns
```

```python
# Top 5 products by total sales amount
top_products = df_sales.groupby("product_id")["total_amount"].sum().nlargest(5)
print(top_products)
```

```python
# Merge with product details to get product names
top_products_details = top_products.reset_index().merge(df_products, on="product_id")
```

```python
print(top_products_details[["product_name", "category", "total_amount"]])
```

```python
# Total sales by city
sales_by_city = df_sales.groupby("city")["total_amount"].sum().sort_values(ascending=False)
print(sales_by_city)
```

```python
# Average delivery time by city
avg_delivery_time_by_city = df_sales.groupby("city")["delivery_time_mins"].mean().sort_values()
print(avg_delivery_time_by_city)
```

```python
df_sales.columns
```

```python
# Sales trend over time (e.g., monthly sales)
df_sales["month"] = df_sales["order_date"].dt.to_period("M")
monthly_sales = df_sales.groupby("month")["total_amount"].sum()
print(monthly_sales)
```

```python
# Sales by product category
sales_by_category = df_sales.merge(df_products,on="product_id")
sales_by_category = sales_by_category.groupby("category")["total_amount"].sum().sort_values(ascending=False)
print(sales_by_category)
```

# **Data Visualization with Matplotlib**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Plot 1: Top 5 Products by Sales Amount
plt.figure(figsize=(12, 7))
sns.barplot(x="product_name", y="total_amount", data=top_products_details,
palette="viridis")
plt.title("Top 5 Products by Total Sales Amount")
plt.xlabel("Product Name")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Total Sales by City
plt.figure(figsize=(12, 7))
sns.barplot(x=sales_by_city.index, y=sales_by_city.values, palette="magma")
plt.title("Total Sales by City")
plt.xlabel("City")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Plot 3: Monthly Sales Trend
plt.figure(figsize=(14, 7))
monthly_sales.plot(kind="line", marker="o", color="#3498DB")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales Amount")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot 4: Sales by Product Category (Pie Chart)
plt.figure(figsize=(10, 10))
plt.pie(sales_by_category, labels=sales_by_category.index,
autopct="%1.1f%%", startangle=140, colors=sns.color_palette("pastel"))
plt.title("Total Sales by Product Category")
plt.axis("equal") # Equal aspect ratio ensures that pie is drawn as a

plt.show()
```

```python
# Plot 5: Distribution of Delivery Times
plt.figure(figsize=(10, 6))
sns.histplot(df_sales["delivery_time_mins"], bins=20, kde=True,
color="#2ECC71")
plt.title("Distribution of Delivery Times")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Delivery Status Count
delivery_counts = df_sales["delivery_status"].value_counts()

plt.figure(figsize=(10,6))
delivery_counts.plot(kind="bar")

plt.title("Delivery Status Distribution")
plt.xlabel("Delivery Status")
plt.ylabel("Number of Orders")

plt.show()
```

```python
#Top Cities by Number of Orders
top_cities_orders = df_sales["city"].value_counts().head(10)

plt.figure(figsize=(10,6))
top_cities_orders.plot(kind="bar")

plt.title("Top Cities by Number of Orders")
plt.xlabel("City")
plt.ylabel("Number of Orders")

plt.show()
```

```python
#Average Order Value by City
avg_order_city = df_sales.groupby("city")["total_amount"].mean().sort_values(ascending=False)

plt.figure(figsize=(12,6))
avg_order_city.plot(kind="bar")

plt.title("Average Order Value by City")
plt.xlabel("City")
plt.ylabel("Average Order Amount")

plt.show()
```

```python
#Quantity Distribution
plt.figure(figsize=(10,6))

plt.hist(df_sales["quantity"], bins=20)

plt.title("Quantity Distribution")
plt.xlabel("Quantity")
plt.ylabel("Frequency")

plt.show()
```

```python
#Daily Sales Trend
df_sales["date"] = df_sales["order_date"].dt.date

daily_sales = df_sales.groupby("date")["total_amount"].sum()

plt.figure(figsize=(14,6))
daily_sales.plot()

plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Total Sales")

plt.show()
```
